```python
#在起飞前多久购买分析
import pandas as pd

products=pd.read_csv('G:\data\data_fam_products.csv') #票价及售量

sale_time = pd.DataFrame(columns = ['RD0', 'RD1', 'RD2', 'RD4', 'RD6','RD7', 'RD8', 'RD11', 'RD14','RD16', 'RD18', 'RD21', 'RD24', 'RD28', 'RD33', 'RD39','RD44', 'RD49', 'RD59', 'RD69', 'RD79', 'RD89', 'RD99', 'RD119', 'RD139', 'RD159', 'RD179', 'RD209', 'RD239', 'RD269', 'RD330'])

sale_time.loc[0]=0
sale_time.loc[1]=1
sale_time
```

```python
import re
for i in range(0,31):
    a = products.iloc[:, [i+7]].sum()
    sale_time.iloc[0,i]=a[0]
    if i == 0:
        sale_time.iloc[1,i] = sale_time.iloc[0,i]
    else:    
        b = re.findall(r"\d+",products.columns[i+6])
        c = re.findall(r"\d+",products.columns[i+7])
        b = int(b[0])
        c = int(c[0])
        sale_time.iloc[1,i] = sale_time.iloc[0,i]/(c-b)

sale_time.index=['sum', 'average']

sale_time
```

```python
sale_time.to_csv('sale_time.csv')
```

```python
products['sale']=products.loc[:,'RD0':'RD330'].sum(axis = 1) 

products.sort_values(['Flight1', 'Flight2', 'Flight3', 'Class'], inplace=True, ascending=True) 

products.reset_index(drop=True,inplace=True)

products
```

```python
fare_sale = pd.DataFrame(columns = ['Flight1', 'Flight2', 'Flight3', 'B_class_fare', 'B_class_sales', 'G_class_fare', 'G_class_sales', 'H_class_fare','H_class_sales', 'K_class_fare','K_class_sales', 'L_class_fare','L_class_sales', 'M_class_fare', 'M_class_sales', 'Q_class_fare', 'Q_class_sales', 'S_class_fare','S_class_sales', 'T_class_fare','T_class_sales', 'V_class_fare','V_class_sales', 'Y_class_fare','Y_class_sales', 'Sales'])  
fare_sale
```

```python
segment_sale = products.groupby(['Flight1', 'Flight2', 'Flight3'], as_index=False)['sale'].sum()  

for i in range(0,len(segment_sale)):
    fare_sale.loc[i,'Flight1'] = segment_sale.Flight1[i]
    fare_sale.loc[i,'Flight2'] = segment_sale.Flight2[i]
    fare_sale.loc[i,'Flight3'] = segment_sale.Flight3[i]
    fare_sale.loc[i,'Sales'] = segment_sale.sale[i]

fare_sale
```

```python
for i in range(0,len(products)):
    if (i%11 == 0):
        fare_sale.B_class_fare[i//11] = products.Fare[i]
        fare_sale.B_class_sales[i//11] = products.sale[i]
    if (i%11 == 1):
        fare_sale.G_class_fare[i//11] = products.Fare[i]
        fare_sale.G_class_sales[i//11] = products.sale[i]
    if (i%11 == 2):
        fare_sale.H_class_fare[i//11] = products.Fare[i]
        fare_sale.H_class_sales[i//11] = products.sale[i]    
    if (i%11 == 3):
        fare_sale.K_class_fare[i//11] = products.Fare[i]
        fare_sale.K_class_sales[i//11] = products.sale[i]
    if (i%11 == 4):
        fare_sale.L_class_fare[i//11] = products.Fare[i]
        fare_sale.L_class_sales[i//11] = products.sale[i]
    if (i%11 == 5):
        fare_sale.M_class_fare[i//11] = products.Fare[i]
        fare_sale.M_class_sales[i//11] = products.sale[i]
    if (i%11 == 6):
        fare_sale.Q_class_fare[i//11] = products.Fare[i]
        fare_sale.Q_class_sales[i//11] = products.sale[i]
    if (i%11 == 7):
        fare_sale.S_class_fare[i//11] = products.Fare[i]
        fare_sale.S_class_sales[i//11] = products.sale[i]
    if (i%11 == 8):
        fare_sale.T_class_fare[i//11] = products.Fare[i]
        fare_sale.T_class_sales[i//11] = products.sale[i]
    if (i%11 == 9):
        fare_sale.V_class_fare[i//11] = products.Fare[i]
        fare_sale.V_class_sales[i//11] = products.sale[i]
    if (i%11 == 10):
        fare_sale.Y_class_fare[i//11] = products.Fare[i]
        fare_sale.Y_class_sales[i//11] = products.sale[i]
```

```python
fare_sale.to_csv("fare_sale.csv")
```
